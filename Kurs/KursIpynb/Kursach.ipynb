```python
def task8(t, n, a):
    for i in range(t):
        x = n[i]
        y = a[i].split(' ')
        cnt = [0 for j in range(x+1)]
        for j in range(x):
            cnt[int(y[j])] += 1
        ans = 0
        for j in range(x):
            summ = 0
            for k in range(j,x):
                summ += int(y[k])
                if j == k:
                    continue
                if summ <= x:
                    ans += cnt[summ]
                    cnt[summ] = 0
        print(ans)
t = 5
n = [9, 3, 5, 8, 1]
a = ["3 1 4 1 5 9 2 6 5", "1 1 2", "1 1 1 1 1", "8 7 6 5 4 3 2 1", "1"]
task8(t, n, a)
```

```python
def countPaths(n):

    zB = 1


    zADC = 0

    # –ù–∞ –∫–∞–∂–¥–æ–º —à–∞–≥–µ zADC —É–º–Ω–æ–∂–∞–µ—Ç—Å—è –Ω–∞ 2(2 —Å–æ—Å—Ç–æ—è–Ω–∏—è) –∏ –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è zB, –ø–æ—Å–∫–æ–ª—å–∫—É zB -
    # —ç—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—É—Ç–µ–π –Ω–∞ —à–∞–≥–µ n - 1, –∫–æ—Ç–æ—Ä–æ–µ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è 2 —Å–æ—Å—Ç–æ—è–Ω–∏–π.
    for i in range(1, n + 1):

        nzB = zADC * 3

        nzADC = (zADC * 2 + zB)


        zB = nzB
        zADC = nzADC


    return zB



n = 4
print(countPaths(n))

```
